```python
import pandas as pd
df=pd.read_excel(r"C:\Users\Shreyasi\Downloads\IPL-2024.xlsx")
matchlist=df.values.tolist()
print(matchlist)
```

```python
##no of matches played
len(matchlist)
```

```python
##maxtarget
maxtarget=-1
for row in matchlist:
    target=row[5]
    if maxtarget<target:
        maxtarget=target
print(maxtarget)
        
       
```

```python
##average run target
avgtarget=0
sum=0
count=0
for row in matchlist:
    sum=sum+row[5]
    count=count+1
avgtarget=sum/count
print(avgtarget)
```

```python
##no of matches above average run target
avgtarget=0
sum=0
count=0
for row in matchlist:
    sum=sum+row[5]
    count=count+1
avgtarget=sum/count
ccount=0
for row in matchlist:
    if row[5]>avgtarget:
        ccount=ccount+1
print("NUMBER OF MATCHES ABOVE AVERAGE MATCH TARGET:",ccount)
```

```python
##how many cities were venues?
venuelist=[]
for row in matchlist:
    venue=row[0]
    if not(venue in venuelist):
        venuelist=venuelist+[venue]
print(venuelist)
```

```python
len(venuelist)
```

```python
##venues with frequencies
venuedict={}
for row in matchlist:
    venue=row[0]
    if venue in venuedict:
        venuedict[venue]=venuedict[venue]+1
    else:
        venuedict[venue]=1
print(venuedict)
```

```python
##which team played as Team 1 in maximum cities?
teamvenuedict={}
for row in matchlist:
    venue=row[0]
    team1=row[1]
    if team1 in teamvenuedict:
        if not (venue in teamvenuedict[team1]):
            teamvenuedict[team1]=teamvenuedict[team1]+[venue]
    else:
            teamvenuedict[team1]=[venue]
print(teamvenuedict)
            
```

```python
maxteam=""
maxvenues=0
for team in teamvenuedict:
    if len(teamvenuedict[team])>maxvenues:
        maxvenues=len(teamvenuedict[team])
        maxteam=team
```

```python
##is winning the toss an advantage?
count=0
countt=0
for row in matchlist:
    count=count+1
    if row[3]==row[4]:
        countt=countt+1
if(countt/count)>0.5:
    print("Winning the toss is an advantage")
else:
    print("Winning the toss is not an advantage")

```
